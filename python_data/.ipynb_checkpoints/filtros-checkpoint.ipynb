```python
# filtrando dados de um data frame

import pandas as pd
```

```python
df = pd.read_csv('./data/users.csv')
```

```python
df.head()
```

```python
df_cols = df[['firstname','email', 'profession']]
df_cols.head()
```

```python
df_30_plus = df[ df['age'] > 30]
df_30_plus.head()
```

```python
df_emprego_idade = df[ (df['age'] >= 30) & (df['profession'] == 'firefighter')]
df_emprego_idade.head()
```

```python
df_emailbr = df[ df['email'].str.contains('.br') | df['email2'].str.contains('.br')]
df_emailbr.head()
```
